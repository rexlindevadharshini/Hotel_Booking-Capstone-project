12. babies : Number of babies

16. distribution_channel : Booking distribution channel
17. is_repeated_guest : Value indicates if the booking name was from a repeated guest (1) or not(0)
18. previous_cancellations : Number of previous bookings that were cancelled by the customer prior to the current booking
19. previous_bookings : Number of previous bookings that were not cancelled by the customer prior to the current booking
20. reserved_room_type : Code of room type reserved
21. assigned_room_type : Code of room type assigned for booking
22. booking_changes : Number of changes made to the booking
23. deposit_type : Indication on if the customer made a deposit to guarantee booking
24. agent : ID of the travel agency that made booking
25. company : ID of the company that made booking
26. days_in_waiting_list : Number of days the booking was  in the waiting list  before it was confirmed to the customer
27. customer_type : Type of booking ,assuming one of four categories
28. adr : Average Daily Rate as defined by dividing the sum of all lodging transction by the total number of staying nights
29. required_car_parking_spaces : Number of car parking spaces required by the customer
30. total_of_special_requests : Number of special request made by customer
31. reservation_status : Reservation last status

## 3. ***Data Wrangling***

### Data cleaning


````python
# treating the missing values
missing_values[0:4]
````

````python
# drop the company column which has more missing values
Hotel_booking.drop(['company'],axis=1,inplace=True)
````

````python
# replace null values with 0 in agent column
Hotel_booking['agent'].fillna(0,inplace=True)
````

````python
# replace null values with 0 in children column
Hotel_booking['children'].fillna(0,inplace=True)
````

````python
# replace null values with 'other' in country
Hotel_booking['country'].fillna('Other',inplace=True)
````

````python
# recheck for null values
Hotel_booking.isna().sum()
````

````python
Hotel_booking.info()
````

````python
# change datatype for children and agent column
Hotel_booking[['children','agent']]= Hotel_booking[['children','agent']].astype('int64')
````

Adding new column

````python
#  create new column for total guests
Hotel_booking['total_guests'] = Hotel_booking['adults'] + Hotel_booking['children'] +Hotel_booking['babies']
````

````python
# create new column for total stays in night
Hotel_booking['total_stays_in_night'] = Hotel_booking['stays_in_weekend_nights'] + Hotel_booking['stays_in_week_nights']
````

````python
# understanding purpose make changes in is_canceled column
Hotel_booking['is_canceled'] = Hotel_booking['is_canceled'].replace([0,1],['not canceled','canceled'])
````

````python
# same for is_repeated_guest column
Hotel_booking['is_repeated_guest'] = Hotel_booking['is_repeated_guest'].replace([0,1],['not repeated','repeated'])
````

#### Chart - 5

````python
# count the no.of bookings in month wise
booking_by_months = Hotel_booking.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':'counts'})

# create a month order list
months=['January','February','March','April','May','June','July','August','September','October','November','December']

# map the above order list without changing its values
booking_by_months['arrival_date_month'] = pd.Categorical(booking_by_months['arrival_date_month'],categories=months,ordered =True)

# sort the values
booking_by_months=booking_by_months.sort_values('arrival_date_month')

booking_by_months
````

````python
# Chart - 5 visualization code
# set plot size and plot lineplot

plt.figure(figsize=(18,6))
sns.lineplot(x=booking_by_months['arrival_date_month'],y=booking_by_months['counts'])
plt.title('Number of bookings over each month')
plt.xlabel('Month')
plt.ylabel('Number of bookings')
plt.show;
````

##### 1. Why did you pick the specific chart?

To know the number of bookings has made over each month.

##### 2. What is/are the insight(s) found from the chart?

The booking has highly raised between July and August.

##### 3. Will the gained insights help creating a positive business impact?
